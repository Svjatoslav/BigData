```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
mu, sigma = 0, 1  # Среднее значение и стандартное отклонение

# Генерируем 195 точек из нормального распределения
data = np.random.normal(mu, sigma, 195)

# Визуализируем данные
def get_hist(data):
    plt.hist(data, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')
    plt.title('Гистограмма нормального распределения')
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.show()
get_hist(data)
```

```python
modern_data = np.append(data, [5,-4,3.3,2.99,-3])
get_hist(modern_data)
```

```python
modern_data_sorted = sorted(modern_data)
```

```python
df = pd.DataFrame(data=modern_data_sorted, columns=['Data'])
```

```python
df['Mean'] = np.mean(modern_data_sorted)
```

```python
df['Deviantion'] = abs(df['Data'] - df['Mean'])
```

```python
df
```

```python
fig = plt.figure(figsize=(10,10))
sns.boxplot(data=df['Data'])

# Добавляем линии для 1, 2 и 3 стандартных отклонений
plt.axhline(np.mean(data) - np.std(data), color='red', linestyle='--', label='-1σ')
plt.axhline(np.mean(data) + np.std(data), color='red', linestyle='--', label='+1σ')

plt.axhline(np.mean(data) - 2 * np.std(data), color='green', linestyle='--', label='-2σ')
plt.axhline(np.mean(data) + 2 * np.std(data), color='green', linestyle='--', label='+2σ')

plt.axhline(np.mean(data) - 3 * np.std(data), color='blue', linestyle='--', label='-3σ')
plt.axhline(np.mean(data) + 3 * np.std(data), color='blue', linestyle='--', label='+3σ')

# Отображаем легенду
plt.legend()

# Показываем график
plt.show()
```
