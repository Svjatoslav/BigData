```python
def get_x_y():
    x= [x for x in range(1,501)]
    y = [0.5*math.sin(k*0.1)+np.random.normal(0,1,1) for k in range(1,501)]
    return x,np.concatenate(y, axis = 0)


def get_real_trend():
    return  [0.5*math.sin(k*0.1) for k in range(1,501)]
```

```python
x, y = get_x_y()
plt.plot(x,y)
```

```python
def get_exp_mov_averange(alpha, x, y):
    arr = [(y[0]+y[1])/2] 
    for i in range(1,len(y)):
        new_val = alpha*y[i] + (1-alpha)*arr[i-1]
        arr.append(new_val)
    return arr
    
```

```python
alpha = 0.1
x,y = get_x_y()
arr = get_exp_mov_averange(alpha, x, y)
```

```python
x, y = get_x_y()
X = np.fft.fft(y)
fs = 10
amplitude_spectrum = np.abs(X)

main_frequency_index = np.argmax(amplitude_spectrum)
main_frequency = main_frequency_index * fs / len(x)

print("Main Frequency:", main_frequency, "Hz")
fig = plt.figure(figsize=(15,5))
plt.plot(amplitude_spectrum)
plt.title("Amplitude Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

```python
def get_noise(alpha,x,y):
#     x,y = get_x_y()
    arr = get_exp_mov_averange(alpha,x,y)
    noise = y - arr
    return noise
```

```python
def countof_turningpoints(lst):
    dx = np.diff(lst,axis=0)
    return np.sum(dx[1:] * dx[:-1] < 0)
```

```python
def count_P(arr):
    count = 0
    for i in range(len(arr)):
        for j in range(i):
            if arr[i] > arr[j]:
                count +=1 
    return count 

def get_Kendal(alpha):
    n = 500
    x,y= get_x_y()
    noise = get_noise(alpha, x, y)
    P = count_P(noise)
    return 4*P/(n*(n-1)) -1

def get_dispersion_Kendal(n =500):
    return sqrt(2*(2*n+5)/(9*n*(n-1)))
```
